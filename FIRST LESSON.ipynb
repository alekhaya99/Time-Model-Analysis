# Quantile- Quantile Plot / QQ- Plot

#### The QQ plot is used to determine whether a data set is distributed in a certain way. Usally showcase how the data fits the normal distribution.

```python
import scipy.stats
import pylab
```

```python
scipy.stats.probplot(df_com.spx,plot=pylab)
pylab.show()
```
